```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
import seaborn as sns
```

```python
df = pd.read_csv("../data/robot_obstacle.csv", header = 1, encoding = 'EUC-KR')
df.head()
```

#### *업종별 데이터프레임 추출

```python
df_jobtype = df.iloc[1:8, 1:]
df_jobtype = df_jobtype.set_index("특성별(2)")
df_jobtype
```

```python
plt.figure(figsize=(30, 8))
ax = sns.heatmap(df_jobtype.iloc[:, 1:], annot=True, fmt='.1f', linewidths=0.7, annot_kws={"size": 20})
ax.set(xlabel="사유", ylabel="업종별")
ax.xaxis.tick_top()
plt.show(ax)
```

##### 위 결과를 통해 '업종별 분류'에서 '초기투자 비용의 부담' 및 '전문인력 부족'이 걸림돌이 된다는 사실을 알 수 있다.
