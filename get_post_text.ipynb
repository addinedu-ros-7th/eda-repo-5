```python
from bs4 import BeautifulSoup
import requests
import mysql.connector
```

```python
conn = mysql.connector.connect(
    host = "*****",
    port = "****",
    user = "****",
    password = "****",
    database = "****"
)

cursor = conn.cursor(buffered=True)
```

```python
cursor.execute("select post_url from recruit")
urls = cursor.fetchall()

for url in urls:
    url = url[0]
    print(url)
    res = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
    soup = BeautifulSoup(res.text, "html.parser")

    #comp_sid
    jv_header = soup.find("div", class_="jv_header")
    if jv_header is None:
        continue
    
    comp_sid = jv_header.find("div", class_="title_inner").find("button", title="관심기업 등록").get("csn")
    
    # comp_form
    info = soup.find("div", class_="wrap_info")
    if info:            
        for dl in info.find("div", class_="info_area").find_all("dl"):
            dt_text = dl.find("dt").get_text(strip=True)
            if "기업형태" in dt_text:
                comp_form = dl.find("dd").get("title")
                cursor.execute("update company set comp_form=%s where comp_sid=%s",
                               (comp_form, comp_sid))

    content = soup.find("div", class_="user_content")
    post_txt = content.get_text(strip=True)
    cursor.execute("update recruit set post_txt=%s where post_url=%s", 
                   (post_txt, url))
    print(comp_form)
    print(post_txt)
```

```python
conn.commit()
```
