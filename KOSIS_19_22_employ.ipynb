```python
df_business_all = pd.DataFrame(data_business_all)
```

```python
df_business_all
```

```python
# 데이터 변환: 연도별 종사자 수
df_business_all_melted = pd.melt(df_business_all, id_vars='업종', value_vars=['2019년 종사자수', '2020년 종사자수', '2021년 종사자수', '2022년 종사자수'], 
                    var_name='연도', value_name='종사자수')

# 연도에서 '년 종사자수' 제거
df_business_all_melted['연도'] = df_business_all_melted['연도'].str.replace('년 종사자수', '')

# 막대 그래프 생성
plt.figure(figsize=(14, 8))
bar_plot = sns.barplot(x='업종', y='종사자수', hue='연도', data=df_business_all_melted, palette='Set3')

# 각 업종별로 막대 그래프 위에 종사자 수 텍스트 추가
for bar in bar_plot.patches:
    height = bar.get_height()
    x_position = bar.get_x() + bar.get_width() / 2  # 막대의 중앙 x좌표

    # 5자리 숫자인 경우와 그렇지 않은 경우의 간격 조정
    if height >= 10000:
        plt.text(x_position, height + 300, int(height), ha='center', va='bottom', fontsize=8)
    else:
        plt.text(x_position, height + 200, int(height), ha='center', va='bottom', fontsize=10)

for idx, 업종 in enumerate(df_business_all['업종']):
    df_filtered = df_business_all[df_business_all['업종'] == 업종].melt(id_vars='업종', 
        value_vars=['2019년 종사자수', '2020년 종사자수', '2021년 종사자수', '2022년 종사자수'], 
        var_name='연도', value_name='종사자수')

    # x축 위치: 업종의 중앙 위치를 가져옴
    x_positions = [idx + i * 0.2 - 0.3 for i in range(4)]  # 각 연도의 x 위치를 설정

    # 꺾은선 그래프 그리기 (막대 그래프 위에)
    plt.plot(x_positions, df_filtered['종사자수'], marker='o', color='blue')

# 그래프 제목 및 레이블 설정
plt.title('업종별 종사자 수 (2019년 ~ 2022년)', fontsize=16)
plt.xlabel('업종', fontsize=12)
plt.ylabel('종사자수', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# 그래프 출력
plt.show()
```
