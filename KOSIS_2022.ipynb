```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
import seaborn as sns
```

```python
df = pd.read_csv("../data/company.csv", header=1, encoding = 'EUC-KR')
df
```

```python
df_jobtype = df_jobtype = df.iloc[1:8, 1:]
df_jobtype = df_jobtype.set_index("특성별(2)")
df_jobtype
```

```python

# 막대 그래프 생성
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x='특성별(2)', y='사업체수 (개)', data=df_jobtype, palette='Set3')
# 그래프 제목 및 레이블 설정
plt.title('업종별 사업체 수', fontsize=16)
plt.xlabel('업종', fontsize=12)
plt.ylabel('사업체 수', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
# 그래프 출력
plt.show()
```
