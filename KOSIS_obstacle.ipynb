```python
class Convert:
    def __init__(self, df, col_name):
        self.df = df
        self.col_name = col_name

    def change_name(self):
        self.df.rename(columns={
            self.df.columns[i]: self.col_name[i] for i in range(9)
        }, inplace=True)
    
    def change_zero(self):
        for name in self.col_name:
            self.df.loc[:, name] = self.df[name].str.replace('-', '0')

    
    def change_type(self):
        self.df = self.df.astype(dtype={self.col_name[0]:"float",
                                        self.col_name[1]:"float",
                                        self.col_name[2]:"float",
                                        self.col_name[3]:"float",
                                        self.col_name[4]:"float",
                                        self.col_name[5]:"float",
                                        self.col_name[6]:"float",
                                        self.col_name[7]:"float",
                                        self.col_name[8]:"float"},
                                        errors="raise",
                                        copy=False)
```
